```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "Personal_Monthly_Expenditure_2021.xlsx"
df = pd.read_excel(file_path)
```

```python
print("Dataset Preview:")
print(df.head())
```

```python
print("\nDataset Info:")
print(df.info())
```

```python
print("\nTotal Expenditure by Category:")
print(df.sum(numeric_only=True))
```

```python
df.set_index("Month", inplace=True)   # Assuming there is a "Month" column
df.plot(kind="bar", figsize=(12,6))
plt.title("Monthly Expenditure by Category (2021)")
plt.ylabel("Amount Spent")
plt.show()
```

```python
while True:
    user_input = input("\nEnter a month or 'exit' to quit: ").capitalize()
    if user_input.lower() == "exit":
        break
    if user_input in df.index:
        print(f"\nExpenditure in {user_input}:")
        print(df.loc[user_input])
    else:
        print("Invalid month. Try again.")
```
